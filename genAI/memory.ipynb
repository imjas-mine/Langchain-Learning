## agent with memory

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
```

```python
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver

```

```python
agent=create_agent(
    model="gpt-5",
    checkpointer=InMemorySaver()
)
    
    
```

```python
config={"configurable":{"thread_id":"test_1"}}
```

```python
from langchain_core.messages import HumanMessage, SystemMessage
response=agent.invoke({"messages":[HumanMessage(content="Hi my name is Jasmine")]},config=config)
print(response)
```

```python
response=agent.invoke({"messages":[HumanMessage(content="Hi what's my name?")]},config=config)
print(response)
```

```python
config1={"configurable":{"thread_id":"test_2"}}
response=agent.invoke({"messages":[HumanMessage(content="Hi my name is Alex")]},config=config1)
print(response)
```

```python
response=agent.invoke({"messages":[HumanMessage(content="Hi what's my name?")]},config=config1)
print(response)
```
